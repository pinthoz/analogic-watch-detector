```python
import pandas as pd

file_path = r'C:\Users\anoca\Documents\GitHub\analogic-watch-detector\ground_truths/watch_times.csv'


original_df = pd.read_csv(file_path)


original_df.head(), original_df.info()
```

```python
original_df
```

```python

rotations = [-90, 180, -270]
new_watches = [
    f"{row['Watch']}_rotated_{rotation}"
    for _, row in original_df.iterrows()
    for rotation in rotations
]


new_data = []

for new_watch in new_watches:
    original_watch = new_watch.split("_rotated_")[0]  
    original_time = original_df.loc[original_df["Watch"] == original_watch, "Time"].values
    if len(original_time) > 0:
        new_data.append({"Watch": new_watch, "Time": original_time[0]})

rotated_df = pd.DataFrame(new_data)


final_df = pd.concat([original_df, rotated_df], ignore_index=True)
```

```python
output_path = r'C:\Users\anoca\Documents\GitHub\analogic-watch-detector\ground_truths/ground_truths.csv'
final_df.to_csv(output_path, index=False)
```
